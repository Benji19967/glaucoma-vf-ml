# Notebook to explore the GRAPE datasets

```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import json
```

```python
GRAPE_DIR = Path("../data") / "GRAPE"
CFPS_DIR = GRAPE_DIR / "CFPs"
ROI_IMAGES_DIR = GRAPE_DIR / "ROI images"
ANNOTATED_IMAGES_DIR = GRAPE_DIR / "Annotated Images"
COORDINATES_DIR = GRAPE_DIR / "json"

IMG_NAME = "1_OD_1.jpg"
JSON_NAME = "1_OD_1.json"
```

### CFPs (Color fundus photography)

Color fundus photography is a non-invasive diagnostic technique that uses a specialized camera to take **high-resolution, full-color images of the retina, optic disc, and macula**

```python
img = Image.open(str(CFPS_DIR / IMG_NAME))
plt.imshow(img)
```

### ROI Images (Region of Interest)

For AI-based model to identify the characteristics more facilely, ROI, a square area around optic dsic (OD), was clipped by one experienced ophthalmologist.


```python
img = Image.open(str(ROI_IMAGES_DIR / IMG_NAME))
plt.imshow(img)
```

### Annotated Images

To singularize the feature of cup-to-disc ratio (CDR) increasing and OD rim narrowing, OD and optic cup (OC) segmentation on ROI images was performed by one ophthalmologist using the label tool. The red circle is the border of OC and green of OD in the annotated images.

```python
img = Image.open(str(ANNOTATED_IMAGES_DIR / IMG_NAME))
plt.imshow(img)
```

### XY Coordinates of Borders

Gives us the xy-coordinates of green (OD) and red (OC) polygons. Each pixel pair represents one corner of the polygon.

```python
FILENAME = COORDINATES_DIR / JSON_NAME
with open(FILENAME) as f:
    coords = json.load(f)

print(json.dumps(coords["shapes"], indent=4))
```
